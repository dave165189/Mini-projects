```
# question 1 Generate random strings
generate_random_string <- function(length) {
  chars <- c(letters, LETTERS, 0:9)
  paste0(sample(chars, length, replace = TRUE), collapse = "")
}

# Create file with random strings
file_path <- "random_strings.txt"
num_lines <- 1000

# Open the file in write mode
file <- file(file_path, "w")

# Generate and write random strings to the file
for (i in 1:num_lines) {
  random_string <- generate_random_string(10)  # Adjust the length as needed
  writeLines(random_string, file)
}

# Close the file
close(file)

# Print confirmation message
cat("File", file_path, "created with", num_lines, "lines of random strings.")
```

```
# (i) Replace all the values with NA in the dataset defined between [10, 60]. Print the count of number rows having missing values.

# Set the dimensions of the dataset
num_rows <- 100
num_cols <- 30

# Generate random dataset
random_dataset <- matrix(sample(1:200, num_rows*num_cols, replace = TRUE), nrow = num_rows)

# Replace values within the range [10, 60] with NA
random_dataset[random_dataset >= 10 & random_dataset <= 60] <- NA

# Count rows with missing values
rows_with_missing <- rowSums(is.na(random_dataset))
num_rows_with_missing <- sum(rows_with_missing > 0)

# Print the count of rows with missing values
cat("Number of rows with missing values:", num_rows_with_missing)
```

```
# (ii) Replace all the NA values with the average of the column value.

# Set the dimensions of the dataset
num_rows <- 100
num_cols <- 30

# Generate random dataset
random_dataset <- matrix(sample(1:200, num_rows*num_cols, replace = TRUE), nrow = num_rows)

# Replace values within the range [10, 60] with NA
random_dataset[random_dataset >= 10 & random_dataset <= 60] <- NA

# Replace NA values with column averages
column_averages <- colMeans(random_dataset, na.rm = TRUE)
random_dataset[is.na(random_dataset)] <- column_averages[col(random_dataset)[is.na(random_dataset)]]

# Count rows with missing values
rows_with_missing <- rowSums(is.na(random_dataset))
num_rows_with_missing <- sum(rows_with_missing > 0)

# Print the count of rows with missing values
cat("Number of rows with missing values:", num_rows_with_missing)
```

```
# (iii) Find the Pearson correlation among all the columns and plot heat map. Also select those columns having correlation <=0.7

# Set the dimensions of the dataset
num_rows <- 100
num_cols <- 30

# Generate random dataset
random_dataset <- matrix(sample(1:200, num_rows*num_cols, replace = TRUE), nrow = num_rows)

# Compute Pearson correlation matrix
cor_matrix <- cor(random_dataset, use = "pairwise.complete.obs")

# Plot heatmap of correlation matrix
heatmap(cor_matrix, col = colorRampPalette(c("blue", "white", "red"))(100))

# Select columns with correlation <= 0.7
selected_columns <- which(apply(cor_matrix, 2, function(x) any(x <= 0.7)))

# Print the selected columns
cat("Selected columns with correlation <= 0.7:", selected_columns)
```

```
# (iv) Normalize all the values in the dataset between 0 and 10

# Set the dimensions of the dataset
num_rows <- 100
num_cols <- 30

# Generate random dataset
random_dataset <- matrix(sample(1:200, num_rows*num_cols, replace = TRUE), nrow = num_rows)

# Normalize values between 0 and 10
normalized_dataset <- apply(random_dataset, 2, function(x) {
  min_val <- min(x)
  max_val <- max(x)
  normalized <- (x - min_val) / (max_val - min_val) * 10
  return(normalized)
})

# Print the normalized dataset
print(normalized_dataset)
```

```
# i) k mean clustering

# Set the dimensions of the dataset
num_rows <- 500
num_cols <- 10

# Generate random dataset
random_dataset <- matrix(nrow = num_rows, ncol = num_cols)

# Set ranges for each column
column_ranges <- list(
  c(-10, 10), c(-10, 10), c(-10, 10), c(-10, 10),
  c(10, 20), c(10, 20), c(10, 20), c(10, 20),
  c(-100, 100), c(-100, 100)
)

# Generate random values for each column within the specified ranges
for (i in 1:num_cols) {
  random_dataset[, i] <- runif(num_rows, column_ranges[[i]][1], column_ranges[[i]][2])
}

# Apply K-means clustering
k <- 3  # Number of clusters
set.seed(123)  # Set seed for reproducibility
kmeans_result <- kmeans(random_dataset, centers = k)

# Print cluster assignments
cluster_assignments <- kmeans_result$cluster
print(cluster_assignments)
```

```
# ii) hierarchical clustering

# Set the dimensions of the dataset
num_rows <- 500
num_cols <- 10

# Generate random dataset
random_dataset <- matrix(nrow = num_rows, ncol = num_cols)

# Set ranges for each column
column_ranges <- list(
  c(-10, 10), c(-10, 10), c(-10, 10), c(-10, 10),
  c(10, 20), c(10, 20), c(10, 20), c(10, 20),
  c(-100, 100), c(-100, 100)
)

# Generate random values for each column within the specified ranges
for (i in 1:num_cols) {
  random_dataset[, i] <- runif(num_rows, column_ranges[[i]][1], column_ranges[[i]][2])
}

# Perform hierarchical clustering
dist_matrix <- dist(random_dataset)  # Calculate the distance matrix
hclust_result <- hclust(dist_matrix)  # Perform hierarchical clustering

# Plot the dendrogram
plot(hclust_result)

# Cut the dendrogram into clusters
k <- 3  # Number of desired clusters
clusters <- cutree(hclust_result, k)

# Print the clusters
print(clusters)
```

```
# Q5. Create a random dataset of 500 rows and 5 columns: All the values are defined between [5,10].Perform the following operations:

# Set the dimensions of the dataset
num_rows <- 500
num_cols <- 5

# Generate random dataset
random_dataset <- matrix(runif(num_rows * num_cols, 5, 10), nrow = num_rows)

# Print the random dataset
print(random_dataset)

```

```
# (i) Perform t-Test on each column

# Set the dimensions of the dataset
num_rows <- 500
num_cols <- 5

# Generate random dataset
random_dataset <- matrix(runif(num_rows * num_cols, 5, 10), nrow = num_rows)

# Perform t-test on each column
for (i in 1:num_cols) {
  col_values <- random_dataset[, i]
  ttest_result <- t.test(col_values)
  
  cat("Column", i, "\n")
  cat("Mean:", mean(col_values), "\n")
  cat("t-value:", ttest_result$statistic, "\n")
  cat("p-value:", ttest_result$p.value, "\n")
  cat("\n")
}
```

```
# (ii) Perform Wilcoxon Signed Rank Test on each column

# Set the dimensions of the dataset
num_rows <- 500
num_cols <- 5

# Generate random dataset
random_dataset <- matrix(runif(num_rows * num_cols, 5, 10), nrow = num_rows)

# Perform Wilcoxon signed-rank test on each column
for (i in 1:num_cols) {
  col_data <- random_dataset[, i]
  test_result <- wilcox.test(col_data, mu = 0, alternative = "two.sided")
  
  # Print the test result for the column
  cat("Column", i, "\n")
  cat("Statistic:", test_result$statistic, "\n")
  cat("p-value:", test_result$p.value, "\n")
  cat("\n")
}
```

```
# (iii) Perform Two Sample t-Test and Wilcoxon Rank Sum Test on Column 3 and Column 4

# Set the dimensions of the dataset
num_rows <- 500
num_cols <- 5

# Generate random dataset
random_dataset <- matrix(runif(num_rows * num_cols, 5, 10), nrow = num_rows)

# Extract column 3 and column 4
column_3 <- random_dataset[, 3]
column_4 <- random_dataset[, 4]

# Perform two-sample t-test
t_test_result <- t.test(column_3, column_4)
cat("Two-sample t-test results:\n")
print(t_test_result)

# Perform Wilcoxon rank sum test
wilcox_test_result <- wilcox.test(column_3, column_4)
cat("Wilcoxon rank sum test results:\n")
print(wilcox_test_result)
```
